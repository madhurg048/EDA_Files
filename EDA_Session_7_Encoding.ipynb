**Advantages**

- Using one hot encoder we will get new columns 

- the new column values at a time only one will be **True** another will be **False**

- Which means new columns are **independant to each other**

- Independant means **no relation**

- Means no **Multi Colinearity**

- Multi collinearity is an important property in Machine Learning

- **Machine Learning models expect the input columns should be independant to each other**

- So one hot encoder gives the no multicollineraity

**Disadvantages**

- The disadvantage is it will increase many columns

- In the above example the total columns become 25510

- This sis happend because of case id

- In real time we will drop the id columns

- Eventhough if we drop still we are getting 30 columns

- More columns means, more time to process

- We reuired more infrastructure (CPU-GPU-TPU)

- This is called **Curse of Dimensionality**
